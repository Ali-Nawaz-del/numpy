## Solving system of linear equations

```python
import numpy as np
A = np.array([
        [4, -3, 1],
        [2, 1, 3],
        [-1, 2, -5]
    ], dtype=np.dtype(float))

b = np.array([-10, 0, 17], dtype=np.dtype(float))

print("Matrix A:")
print(A)
print("\nArray b:")
print(b)
```

```python
x = np.linalg.solve(A,b)
print(x)
```

## Determinant of a matrix

```python
d = np.linalg.det(A)
print(d)
```

## Inverse of a matrix

```python
import numpy as np
A = np.array([
        [1, 2, 3],
        [0, 1, 4],
        [5, 6, 0]
    ], dtype=np.dtype(int))

A_inv = np.linalg.inv(A)
print(A_inv)
```
